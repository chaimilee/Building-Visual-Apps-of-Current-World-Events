```python
#extracting GPE and LOC from each date in news source
def NER_Loc(text,date):
  entity_dict = dict()

  doc = nlp(text)
  locations=[]
  for entity in doc.ents:
    if entity.label_ not in ["GPE","LOC"]:
      pass
    else:
      locations.append(entity.text)
  entity_dict[date]=locations

  return entity_dict
```

```python
list_of_stuff=[]
for k,v in data.items():
  list_of_stuff.append(NER_Loc(v,k))

list_of_stuff2=[]
for k,v in data.items():
  list_of_stuff.append(NER_Loc(v,k))

list_of_stuff3=[]
for k,v in data.items():
  list_of_stuff.append(NER_Loc(v,k))
```

```python
#new list of dict with counts of places
new_list=[]
for element in list_of_stuff2:
  list_of_num=[]
  new_dict={}
  for key,val in element.items():
    my_dict = {i:val.count(i) for i in val}
    for k,v in my_dict.items():
      list_of_num.append((k,v))
  new_dict[key]=list_of_num
  new_list.append(new_dict)
```
