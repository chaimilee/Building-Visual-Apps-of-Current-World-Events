```python
import plotly.graph_objects as go
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.geocoders import Photon
```

```python
df_fox = pd.read_csv('/content/drive/MyDrive/DSCI 550 Team Project/Assignment 3/coordinates_fox.csv')
df_alj = pd.read_csv('/content/drive/MyDrive/DSCI 550 Team Project/Assignment 3/coordinates_aljazeera.csv')
df_cnn = pd.read_csv('/content/drive/MyDrive/DSCI 550 Team Project/Assignment 3/coordinates_cnn.csv')
```

```python
geolocator = Nominatim(user_agent="name_of_your_app")
```

```python
coord1 = df_fox['03-13-22'].tolist()
coord2 = df_alj['3-13-22'].tolist()
coord3 = df_cnn['03-13-22'].tolist()
```

```python
coord_final = coord1 + coord2 + coord3
```

```python
name_count = {}

for element in coord_final:
  # print(element)
  try:
    tup = element[1:-1].split(',')
    # print(tup)
    name_count[tup[3]] = name_count.get(tup[3], 0) + int(tup[2])
  except:
    pass
```

```python
longi=[]
lat=[]
size_list=[]
name=[]
names_seen = []
for element in coord_final:
  # print(element)
  try:
    tup = element[1:-1].split(',')
    while tup[3] not in names_seen:
      lat.append(float(tup[0]))
      longi.append(float(tup[1]))
      size_list.append(name_count[tup[3]])
      name.append(tup[3])
      names_seen.append(tup[3])
  except:
    pass
```

```python
fig = go.Figure(data=go.Scattergeo(
        lon = longi,
        lat = lat,
        mode = 'markers',
        text=name,
        marker = dict(
            size = 10,
            colorscale = 'Reds',
            cmin = 0,
            color = size_list,
            cmax = max(size_list),
            colorbar_title="number of times mentioned"
          )))
          

fig.update_layout(
        title = 'Combined_03-01-22',
        geo_scope='europe',
    )

# fig.write_image("/content/drive/MyDrive/DSCI 550 Team Project/Assignment 3/combined_03-13-22.png")
fig.show()
```
